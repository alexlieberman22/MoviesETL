```python
def clean_movie(movie):
    movie = dict(movie)
    alt_titles = {}
    for key in ['Also known as','Arabic','Cantonese','Chinese','French',
                'Hangul','Hebrew','Hepburn','Japanese','Literally',
                'Mandarin','McCune–Reischauer','Original title','Polish',
                'Revised Romanization','Romanized','Russian',
                'Simplified','Traditional','Yiddish']:
        if key in movie:
            alt_titles[key] = movie[key]
            movie.pop(key)
        
    if len(alt_titles) > 0:
        movie['alt_titles'] = alt_titles
        
    return movie
```

```python
clean_movies = [clean_movie(movie) for movie in wiki_movies]
```

```python
def parse_dollars(s):
    
    # if s is not a string, return NaN
    if type(s) != str:
        return np.nan
    
    # if the number has the form '$###.# million' 
    if re.match(r'\$\s*\d+\.?\d*\s*milli?on', s, flags=re.I):
        # remove what isn't a number
        s = re.sub('\$|\s|[a-zA-Z]', '', s)
        # and multiply the number by a million
        value = float(s) * 10**6
        return value 
    
    # same for a billion
    elif re.match(r'\$\s*\d+\.?\d*\s*billi?on', s, flags=re.I):
        s = re.sub('\$|\s|[a-zA-Z]', '', s)
        value = float(s) * 10**9
    
    # if the number has the form '$###,###,###'
    elif re.match(r'\$\s*\d{1,3}(?:[,\.]\d{3})+(?!\s[mb]illion)', s, flags=re.IGNORECASE):
        # remove what isn't a number
        s = re.sub('\$|,','', s)
        # return the value as a float
        value = float(s)
        return value

    # otherwise, return NaN
    else:
        return np.nan
```

```python
wiki_movie_df_4['box_office'] = box_office.str.extract(f'({form_one}|{form_two})', flags=re.I)[0].apply(parse_dollars)
```
